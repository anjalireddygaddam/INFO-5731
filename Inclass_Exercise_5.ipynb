```
from bs4 import BeautifulSoup
import requests
import pandas as pd
main_heading=[] #creating list to store all the names of the reviews.
body=[] #List to store the description of the reviews.
#rate=[]# To store the rating of the products
for number in range(50):
   html_page=requests.get("https://www.flipkart.com/apple-airpods-pro-wireless-charging-case-active-noise-cancellation-enabled-bluetooth-headset/product-reviews/itm5aa94d4e64bd8?pid=ACCFHYJ8ZU9WDHDW&lid=LSTACCFHYJ8ZU9WDHDWZNRRJJ&marketplace=FLIPKART").text
   soup=BeautifulSoup(html_page,'lxml')  # Instance for the BeautifulSoup library\n",
   review_name=soup.find_all(class_='_2-N8zT') #getting name of the reviews by using class name\n",
   review_description=soup.find_all(class_='t-ZTKy') #getting description of the reviews by class name\n",
   for a, b in zip(review_name, review_description): #iterating the list\n",
      main_heading.append(a.text) #appending the required elements to empty list\n",
      body.append(b.text)
data_frame=pd.DataFrame(list(zip(main_heading, body)), columns=['Review Title','Review description']) #dataframe creation\n",
print("Length of data frame is {0}".format(len(data_frame))) 
data_frame
```

```
data_frame['After Preprocessing'] = data_frame['Review description'].apply(lambda x: " ".join(x.lower() for x in x.split()))
data_frame
```

```
data_frame['After Preprocessing'] = data_frame['After Preprocessing'].str.replace('[^\w\s]','')
data_frame
```

```
data_frame['After Preprocessing'] = data_frame['After Preprocessing'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
```

```
import re
data_frame['After Preprocessing'] = data_frame['After Preprocessing'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))
data_frame
```

```

import nltk
nltk.download()
```

```
from nltk.corpus import stopwords
s = stopwords.words('english')
data_frame['After Preprocessing'] = data_frame['After Preprocessing'].apply(lambda x: " ".join(x for x in x.split() if x not in s))
data_frame
```

```
from nltk.stem import PorterStemmer
stem_ = PorterStemmer()
data_frame['After Preprocessing'].apply(lambda x: " ".join([stem_.stem(word) for word in x.split()]))
```

```
from textblob import Word
import nltk
nltk.download('wordnet')
data_frame['After Preprocessing'] = data_frame['After Preprocessing'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
data_frame
```

```
from nltk.tokenize import word_tokenize
parts_speech= []
for row in data_frame['After Preprocessing']:
  t = word_tokenize(row)
  parts_speech.append(nltk.pos_tag(t))
parts_speech
```

```
Adjective_ = []
Adverb_ = []
Cordinating_Conjunction = []
Sub_ordinating_Conjuction = []
Interjection_ = []
Noun_ = []
Verb_ = []
PersonalPronoun_ = []
predeterminer_ = []
Determiner_= []
```

```
for value in parts_speech:
  Adj_Count = 0
  Adv_Count = 0
  CC_Count = 0
  SC_Count = 0
  I_Count = 0
  Noun_Count = 0
  Verb_Count = 0
  PP_Count = 0
  predeterminer_Count = 0
  Determiner_Count = 0
  for word,tag in value:
    if tag == 'JJ':
      Adj_Count= Adj_Count + 1
    elif tag == 'RB':
      Adv_Count = Adv_Count + 1
    elif tag == 'CC':
      CC_Count = CC_Count + 1
    elif tag == 'UH':
      I_Count = I_Count + 1
    elif tag == 'NN':
      Noun_Count = Noun_Count + 1
    elif tag == 'VR':
      Verb_Count = Verb_Count + 1
    elif tag == 'PRP':
      PP_Count = PP_Count + 1
    elif tag == 'PDT':
      predeterminer_Count = predeterminer_Count + 1
    elif tag == 'DT':
      Determiner_Count = Determiner_Count + 1
    elif tag == 'IN':
      SC_Count = SC_Count + 1
  Adjective_.append(Adj_Count)
  Adverb_.append(Adv_Count)
  Cordinating_Conjunction.append(CC_Count)
  Interjection_.append(I_Count)
  Noun_.append(Noun_Count)
  Verb_.append(Verb_Count)
  PersonalPronoun_.append(PP_Count)
  predeterminer_.append(predeterminer_Count)
  Determiner_.append(Determiner_Count)
  Sub_ordinating_Conjuction.append(SC_Count)
```

```
data_frame['Number of Adjectives'] = Adjective_
data_frame['Number of Adverbs'] = Adverb_
data_frame['Number of Cordinating Conjunctions'] = Cordinating_Conjunction
data_frame['Number of Interjections'] = Interjection_
data_frame['Number of Nouns'] = Noun_
data_frame['Number of Verbs'] = Verb_
data_frame['Number of Personal Pronouns'] = PersonalPronoun_
data_frame['Number of Predeterminers'] = predeterminer_
data_frame['Number of Determiners'] = Determiner_
data_frame['Number of Subordinating Conjuctions'] = Sub_ordinating_Conjuction
data_frame
```

```
import spacy.cli
spacy.cli.download("en_core_web_sm")
```

```
Right_Branching = []
n = spacy.load("en_core_web_sm")
for row in data_frame['After Preprocessing']:
  text_doc = n(row)
  try:
    Right_Branching.append(text_doc[0].n_rights)
  except:
    Right_Branching.append('No')
data_frame['Number of Right Branching Nodes'] = Right_Branching
```
