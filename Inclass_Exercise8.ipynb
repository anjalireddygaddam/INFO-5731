```python
import pandas as pd
data_frame = pd.read_csv('https://raw.githubusercontent.com/anjalireddygaddam/INFO-5731/main/Reviews_Sentimental_Analysis.csv')
data_frame = data_frame.loc[:, ~data_frame.columns.str.contains('^Unnamed')]
```

```python
from nltk.corpus import stopwords
from textblob import TextBlob
data_frame['Review'] = data_frame['Review'].apply(lambda x: " ".join(x.lower() for x in x.split()))
data_frame['Review'] = data_frame['Review'].str.replace('[^\w\s]','')
stop = stopwords.words('english')
data_frame['Review'] = data_frame['Review'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
data_frame['Review'] = data_frame['Review'].apply(lambda x: str(TextBlob(x).correct()))
```

```python
from polyglot.text import Text
sentiment_words = []
for row in data_frame['Review']:
  text = Text(row)
  for x in text.words:
    if x.polarity != 0:
      sentiment_words.append(x)
sentiment_words
```

```python
from collections import Counter
frequency = Counter(sentiment_words)
rank_data_frame = pd.DataFrame(list(frequency.most_common()), columns=['Words', 'Frequencies'])
rank_data_frame.index = list(range(1, len(frequency.most_common())+1))
rank_data_frame
```

```python
# Write your code here
def Encoder(sentiment_analysis):
  senti_encoded = 0
  if sentiment_analysis == 'Positive':
    senti_encoded = 1
  elif sentiment_analysis == 'Negative':
    senti_encoded = 2
  return senti_encoded
```

```python
sentiment_ = []
for sentiment_analysis in data_frame['Sentimental Analysis']:
  sentiment_.append(Encoder(sentiment_analysis))
```

```python
from textblob import TextBlob
text_blob_sentiments_ = []
for row in data_frame['Review']:
  polar = TextBlob(row).sentiment.polarity
  if polar > 0:
    text_blob_sentiments_.append(1)
  elif polar < 0:
    text_blob_sentiments_.append(2)
  elif polar == 0.0:
    text_blob_sentiments_.append(0)
text_blob_data_frame = pd.DataFrame(list(zip(data_frame['Review'], sentiment_, text_blob_sentiments_)), columns = ['Reviews', 'Actual', 'Predicted'])
text_blob_data_frame
```

```python
from sklearn.metrics import f1_score, accuracy_score
textblob_accuracy_ = accuracy_score(text_blob_data_frame['Actual'], text_blob_data_frame['Predicted'])*100
text_blobf1score = f1_score(text_blob_data_frame['Actual'], text_blob_data_frame['Predicted'], average = 'macro')
print("Text Blob Accuracy is {0} and F1 Score is {1}".format(textblob_accuracy_, text_blobf1score))
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
T_Fid_vectorizer = TfidfVectorizer(min_df = 5,
                             max_df = 0.8,
                             sublinear_tf = True,
                             use_idf = True)
```

```python
Train_X, Test_X, Train_Y, Test_Y = train_test_split(data_frame['Review'],data_frame['Sentimental Analysis'],test_size=0.33)
Encoder_ = LabelEncoder()
Train_Y = Encoder_.fit_transform(Train_Y)
Test_Y = Encoder_.fit_transform(Test_Y)
Train_X_vectors = T_Fid_vectorizer.fit_transform(Train_X)
Test_X_vectors = T_Fid_vectorizer.transform(Test_X)
```

```python
from sklearn import svm
from sklearn.metrics import classification_report
svm_model_ = svm.SVC(kernel='linear')
svm_model_.fit(Train_X_vectors, Train_Y)
predict = svm_model_.predict(Test_X_vectors)
reported = classification_report(Test_Y, predict, output_dict=True)
reported
```
